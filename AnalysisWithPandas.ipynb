```python
import pandas as pd
import numpy as np

# ==========================================
# 1. SAMPLE DATA GENERATION
# ==========================================
# We generate a local 'regional_data.csv' to demonstrate
# handling structured business records.
def create_local_dataset():
    data = {
        'region': ['East', 'West', 'North', 'South', 'East', 'West', 'North', 'South'],
        'sales_rep': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank', 'Grace', 'Hank'],
        'revenue': [45000, 52000, 31000, 28000, 61000, 48000, 39000, 32000],
        'transactions': [150, 180, 95, 80, 210, 160, 115, 90],
        'active_status': [True, True, False, True, True, False, True, False]
    }
    df_sample = pd.DataFrame(data)
    df_sample.to_csv('regional_performance.csv', index=False)
    print("Dataset 'regional_performance.csv' initialized.\n")

create_local_dataset()

# ==========================================
# 2. DATA INSPECTION & EXPLORATION
# ==========================================
# Initial assessment of the dataset's structure and health.
df = pd.read_csv('regional_performance.csv')

# Viewing the first 5 rows and summary metadata
print("--- Data Preview ---")
print(df.head())
print("\n--- Structural Information ---")
print(df.info())

# ==========================================
# 3. SORTING & SUBSETTING
# ==========================================
# Reordering the dataset by performance metrics to identify top outliers.
# We sort by revenue in descending order.
df_sorted = df.sort_values('revenue', ascending=False)

# Filtering: Isolating active reps with revenue over 40,000.
high_performers = df[(df['active_status'] == True) & (df['revenue'] > 40000)]

print("\n--- Top Active Performers ---")
print(high_performers[['sales_rep', 'revenue']])

# ==========================================
# 4. DATA TRANSFORMATION & CALCULATIONS
# ==========================================
# Adding a calculated column for 'Average Transaction Value' (ATV).
df['avg_transaction_value'] = df['revenue'] / df['transactions']

# Rounding for presentation
df['avg_transaction_value'] = df['avg_transaction_value'].round(2)

# ==========================================
# 5. SUMMARY STATISTICS & GROUPING
# ==========================================
# Aggregating data by region to compare performance across territories.
regional_summary = df.groupby('region')[['revenue', 'transactions']].sum()

# Adding a count of reps per region
regional_summary['rep_count'] = df.groupby('region')['sales_rep'].count()

print("\n--- Regional Summary Analysis ---")
print(regional_summary)

# ==========================================
# 6. DATA EXPORT
# ==========================================
# Saving the processed and cleaned dataset to a new CSV for reporting.
df_sorted.to_csv('sorted_performance_report.csv', index=False)
print("\nSuccess: 'sorted_performance_report.csv' has been generated.")
```
